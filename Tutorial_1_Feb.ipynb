## 1.
An election will be held next week and, by polling a sample of the voting
population, we are trying to predict whether the Republican or Democratic
candidate will prevail. Which of the following methods of selection is likely to
yield a representative sample?

- (a) Poll all people of voting age attending a college basketball game.
- (b) Poll all people of voting age leaving a fancy midtown restaurant.
- (c) Obtain a copy of the voter registration list, randomly choose 100 names, and question them.
- (d) Use the results of a television call-in poll, in which the station asked its listeners to call in and name their choice.

### Solution:

**(c) Obtain a copy of the voter registration list, randomly choose 100 names, and question them.**

Involves random selection from the voter registration list, which is more likely to reflect the general voting population.

- **(a)** Bias of location
- **(b)** Bias of socioeconomic status
- **(d)** Bias of time, people listening to a specific channel
- **(e)** Not everyone has/had a telephone

## 4
The following table gives the number of commercial airline accidents and fatalities in the United States in the years from 1980 to 1995.
- **(a)** Represent the number of yearly airline accidents in a frequency table.
- **(b)** Give a frequency polygon graph of the number of yearly airline accidents.
- **(c)** Give a cumulative relative frequency plot of the number of yearly airline
accidents.
- **(d)** Find the sample mean of the number of yearly airline accidents.
- **(e)** Find the sample median of the number of yearly airline accidents.
- **(f)** Find the sample mode of the number of yearly airline accidents.
- **(g)** Find the sample standard deviation of the number of yearly airline accidents.


image.png

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airline = {'Year' : [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995],
           'Fatal Accidents' : [0, 4, 4, 4, 1, 4, 2, 4, 3, 11, 6, 4, 4, 1, 4, 2],
           'Fatalities' : [0, 4, 233, 5, 4, 197, 5, 231, 285, 278, 39, 62, 33, 1, 239, 166]}

accidents = pd.DataFrame(airline)
accidents
```

### Solution:

**(a)**

```python
# (a) Frequency Table
frequency = accidents['Fatal Accidents'].value_counts().sort_index()
frequency
```

**(b)**

```python
# (b) Polygon Graph
frequency.plot(kind = 'line')

plt.xlabel('Yearly Airline Accidents')
plt.ylabel('Frequency')
plt.title('Polygon graph of the number of yearly airline accidents')
```

**(c)**

```python
# (c) Cumulative relative frequency plot
relative = accidents['Fatal Accidents'].value_counts(normalize = True).sort_index().cumsum()

relative.plot(kind = 'line')

plt.xlabel('Yearly Airline Accidents')
plt.title('Cumulative relative frequency plot of the number of yearly airline accidents')
```

**(d)**

```python
# (d) Sample Mean
accidents['Fatal Accidents'].mean()
```

**(e)**

```python
# (e) #Sample Median
accidents['Fatal Accidents'].median()
```

**(f)**

```python
# (f) Sample Mode
accidents['Fatal Accidents'].mode()
```

**(g)**

```python
# (g) Sample Standard Deviation
accidents['Fatal Accidents'].std()
```
